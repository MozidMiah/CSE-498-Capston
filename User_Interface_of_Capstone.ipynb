```python
from tkinter import*
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import learning_curve 
import seaborn as sns

root= Tk()
root.title("Squad Selection")
root.config(bg="powderblue")
root.resizable(width= False , height=False)
root.geometry("350x500+200+100")
def naiveBatsman(x):
    
    df = pd.read_csv('Batsman_All - Copy.csv')
    df['rating'] = df['Runs'] / df['Inns'] + df['Runs'] / df['BF'] + df['hundred'] + df['fifty'] / df['Inns'] + df['four']*4 / df['Inns'] + df['six']*6 / df['Inns']
    df['res'] = df['rating'].apply(lambda x: '1' if x >= 20 else '0')
    target=df.res
    inputs=df.drop(['res','Player','Span','rating'],axis='columns')
    X_train, X_test, y_train, y_test= train_test_split(inputs,target,test_size=0.3)
    model = GaussianNB()
    model.fit(X_train, y_train)
    z= model.score(X_test, y_test)
    dm = pd.read_csv('Batsman_Current - Copy.csv')
    dm['rating'] = dm['Runs'] / dm['Inns'] + dm['Runs'] / dm['BF'] + dm['hundred'] + dm['fifty'] / dm['Inns'] + dm['four']*4 / dm['Inns'] + dm['six']*6 / dm['Inns']
    p=dm.drop(['Player','Span','rating'],axis='columns')
    s=dm.sort_values(by='rating', ascending=False)
    k=s.Player
    l=k.head(x)
    df_no_indices = l.to_string(index=False)
    answer= Label(root, text="Batsman", bg="powderblue" )
    answer.pack(pady=1)
    label1= Label(root, text=df_no_indices)
    label1.pack(padx=2, pady=2) 
    
def naiveBolwerfast(y):
    df = pd.read_csv('Boller_All.csv')
    target=df.res
    inputs=df.drop(['res','Player','Span'],axis='columns')
    X_train, X_test, y_train, y_test= train_test_split(inputs,target,test_size=0.3)
    model = GaussianNB()
    model.fit(X_train, y_train)
    w=model.score(X_test, y_test)
    dm = pd.read_csv('Bowller_Cur_fast.csv')
    p=dm.drop(['res','Player','Span','rating'],axis='columns')
    q=dm.res
    s=dm.sort_values(by='rating', ascending=False)
    k=s.Player
    l=k.head(y)
    df_no_indices = l.to_string(index=False)
    answer2= Label(root, text="Fast Bolwer", bg="powderblue" )
    answer2.pack(pady=2)
    label2= Label(root, text=df_no_indices)
    label2.pack(padx=2, pady=2) 

    
def naiveBolwerspin(z):
    df = pd.read_csv('Boller_All.csv')
    target=df.res
    inputs=df.drop(['res','Player','Span'],axis='columns')
    X_train, X_test, y_train, y_test= train_test_split(inputs,target,test_size=0.3)
    model = GaussianNB()
    model.fit(X_train, y_train)
    w=model.score(X_test, y_test)
    dm = pd.read_csv('Bowller_Cur_spin.csv')
    p=dm.drop(['res','Player','Span','rating'],axis='columns')
    q=dm.res
    s=dm.sort_values(by='rating', ascending=False)
    k=s.Player
    l=k.head(z)
    df_no_indices = l.to_string(index=False)
    answer3= Label(root, text="Spin Bolwer", bg="powderblue" )
    answer3.pack(pady=2)
    label3= Label(root, text=df_no_indices)
    label3.pack(padx=2, pady=2) 
    
def number():
    try: 
        a = int(entry1.get())
        b = int(entry2.get())
        c = int(entry3.get())
        naiveBatsman(a)
        naiveBolwerfast(b)
        naiveBolwerspin(c)
                    
    except ValueError:
        answer.config(text="Oops! You Didn't Give Integer Values")
        print("Oops!")
                                
label1= Label(root, text="Number of Batsmans", bg="powderblue")
label1.pack(pady=5)

entry1= Entry(root)
entry1.pack(pady=5)

label2= Label(root, text="Number of Fast Bowlers", bg="powderblue")
label2.pack(pady=5)

entry2= Entry(root)
entry2.pack(pady=5)

label3= Label(root, text="Number of Spin Bowlers", bg="powderblue")
label3.pack(pady=5)

entry3= Entry(root)
entry3.pack(pady=5)

sub_btn = Button(root,text = 'Show',command = number) 
sub_btn.pack(pady=5)

root.mainloop()
```
